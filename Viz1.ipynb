```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# import plotly.plotly as py
# import plotly.graph_objs as go
```

```python
df = pd.read_csv("train.csv", parse_dates=['date'])
df['date'] = pd.to_datetime(df['date'])
```

```python
df['dayofmonth'] = df.date.dt.day
df['dayofyear'] = df.date.dt.dayofyear
df['dayofweek'] = df.date.dt.weekday_name.str.slice(0,3)
df['month'] = df.date.dt.month
df['year'] = df.date.dt.year
df['weekofyear'] = df.date.dt.weekofyear
df['quarter'] = df.date.dt.quarter
```

```python
plt.figure(figsize=(12,5))
plt.title("Distribution of sales - for each item, date and store", fontsize = 20, weight = "bold")
sns.distplot(df['sales']);
plt.savefig("Distribution.png")
```

```python
df = df.drop(['date'], axis =1)
```

```python
year_df = df.groupby('year')['sales'].sum()
y_df = pd.DataFrame(year_df)
y_df = y_df.reset_index()
plt.figure(figsize=(12,5))
plt.title("Sales in each year", fontsize = 20, weight = "bold")
sns.barplot(x = "year", y = "sales", data = y_df, capsize = 0, errwidth = 0);
plt.xlabel('Year', fontsize = 15, weight = "bold");
plt.ylabel('Sales', fontsize = 15, weight = "bold");
plt.savefig("Sales_year.png")
```

```python
mon_df = df.groupby('month')['sales'].sum()
m_df = pd.DataFrame(mon_df)
m_df = m_df.reset_index()
plt.figure(figsize=(12,5))
plt.title("Sales forn each month", fontsize = 20, weight = "bold")
sns.barplot(x = "month", y = "sales", data = m_df, capsize = 0, errwidth = 0);
plt.xlabel('Month', fontsize = 15, weight = "bold");
plt.ylabel('Sales', fontsize = 15, weight = "bold");
plt.savefig("Sales_month.png")
```

```python
dy_df = df.groupby('dayofmonth')['sales'].sum()
d_df = pd.DataFrame(dy_df)
d_df = d_df.reset_index()
plt.figure(figsize=(12,5))
plt.title("Sales for each day in a month", fontsize = 20, weight = "bold")
sns.boxplot(x = "dayofmonth", y = "sales", data = df);
plt.xlabel('Day of the Month', fontsize = 15, weight = "bold");
plt.ylabel('Sales', fontsize = 15, weight = "bold");
plt.savefig("Sales_day.png")
```

```python
wk_df = df.groupby('dayofmonth')['sales'].sum()
w_df = pd.DataFrame(wk_df)
w_df = w_df.reset_index()
plt.figure(figsize=(12,5))
order = ["Sun", "Mon", "Tue", "Wed", "Thu", "Fri", "Sat"]
plt.title("Sales on weekdays", fontsize = 20, weight = "bold")
sns.boxplot(x = "dayofweek", y = "sales", data = df, order = order);
plt.xlabel('Weekday', fontsize = 15, weight = "bold");
plt.ylabel('Sales', fontsize = 15, weight = "bold");
plt.savefig("Sales_weekday.png")
```

```python
store_df = df.groupby('store')['sales'].sum()
plt.figure(figsize=(12,5))
plt.title("Sales in all the store for the period of five years", fontsize = 20, weight = "bold")
plt.plot(store_df);
plt.legend();
plt.xlabel('Store', fontsize = 15, weight = "bold");
plt.ylabel('Sales', fontsize = 15, weight = "bold");
plt.savefig("Sales_store.png");
```

```python
item_df = df.groupby('item')['sales'].sum()
plt.figure(figsize=(12,5))
plt.title("Count of Items Sales for five years", fontsize = 20, weight = "bold")
plt.plot(item_df);
plt.xlabel('Items', fontsize = 15, weight = "bold");
plt.ylabel('Sales', fontsize = 15, weight = "bold");
plt.legend()
plt.savefig("Sales_items.png");
```

```python
plt.figure(figsize=(12,5))
plt.title("Sales for each quarter", fontsize = 20, weight = "bold")
sns.boxplot(x = "quarter", y = "sales", data = df);
plt.xlabel('Quarter', fontsize = 15, weight = "bold");
plt.ylabel('Sales', fontsize = 15, weight = "bold");
plt.savefig("Sales_Quarter.png")
```
